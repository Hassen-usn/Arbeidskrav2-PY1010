```python
#Oppgave 5)
import numpy as np

#definerer funksjon for å regne ut areal og omkrets av trekant med halvsirkel

def fun_beregn(a, b):
    areal1 = 0.5 * np.pi* (a**2) + (b*a)*0.5
    ytreomkrt = 2*(np.pi*a) + (b+b)
    return (areal1, ytreomkrt)

#Tar imot input fra bruker i typ float
aarg = float(input("Skriv radius/høyde for inn agrument a: "))
grunnlinje = float(input("Skriv grunnlinjen for inn argument b: "))




areal, omkrets = fun_beregn(aarg, grunnlinje)  #sender aarg og grunnlinje som argumenter til funksjonen fun_beregn
print("Areal av figur er: ", areal)
print("Omkrets av figur er: ", omkrets)   #Returneres to verdier  areal og omkrets. 

```
